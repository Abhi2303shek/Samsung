```python
import numpy as np
import pandas as pd
```

```python
samsung = pd.read_csv('samsung.csv', sep=',')
samsung
```

```python
samsung.columns
```

```python
samsung[['length', 'width', 'height']] = (samsung['Dimensions'].str.extract(r'([\d.]+)\s*x\s*([\d.]+)\s*x\s*([\d.]+)'))
```

```python
samsung[['soc_core', 'soc_model_name']] = (samsung['SoC'].str.extract(r'^(.*?)\s*\(([^)]+)\)'))
```

```python
samsung['gpu_hertz_mhz'] = samsung['GPU'].str.extract(r'([\d.]+)\s*MHz')
samsung['gpu_name'] = samsung['GPU'].str.replace(r'@.*|[\d.]+\s*MHz', '', regex=True).str.strip()
```

```python
ram_vals = samsung['RAM'].str.extractall(r'(\d+)').unstack()

samsung['ram1_gb'] = ram_vals.iloc[:, 0]
samsung['ram2_gb'] = ram_vals.iloc[:, 1]
samsung['ram_hertz_mhz'] = samsung['RAM'].str.extract(r'(\d+)\s*MHz')
```

```python
rom_vals = samsung['Storage'].str.extractall(r'(\d+)').unstack()
```

```python
samsung['rom1_gb'] = rom_vals.iloc[:, 0]
samsung['rom2_gb'] = rom_vals.iloc[:, 1]
```

```python
cam_vals = samsung['Camera'].str.extractall(r'(\d+)').unstack()
```

```python
samsung['cam1_mp'] = cam_vals.iloc[:, 0]
samsung['cam2_mp'] = cam_vals.iloc[:, 1]
samsung['fps'] = samsung['Camera'].str.extract(r'(\d+)\s*fps')
```

```python
samsung.columns
```

```python
samsung[['battery_capacity_mah', 'battery_type']] = (samsung['Battery'].str.extract(r'(\d+)\s*mAh,\s*(.*)'))
```

```python
col_to_drop = ['SoC', 'Dimensions', 'GPU', 'RAM', 'Storage', 'Camera', 'Battery']
samsung.drop(columns=col_to_drop, inplace=True)
```

```python
samsung.columns
```

```python
samsung
```

## ðŸ”‹ Battery & Design


### Q1. Which devices offer the highest battery capacity per mm of thickness?


```python
samsung['height'] = pd.to_numeric(samsung['height'], errors='coerce')
samsung['battery_capacity_mah'] = pd.to_numeric(
    samsung['battery_capacity_mah'], errors='coerce'
)
```

```python
samsung.assign(
    battery_per_mm=samsung['battery_capacity_mah'] / samsung['height'].astype(float)
).sort_values('battery_per_mm', ascending=False).head(10)
```

### Q2. What is the average battery capacity by device size category (small / medium / large)?


```python
samsung['device_size'] = pd.cut(samsung['length'].astype(float), 
                                bins=[0,200,250,300],
                                labels=['Small', 'Mid','Large'])
```

```python
samsung.groupby('device_size')['battery_capacity_mah'].mean()
```

### Q3. Do thinner devices consistently have lower battery capacities?


```python
samsung[['height','battery_capacity_mah']].corr()
```

### Q4. Which models balance large batteries with compact dimensions best?


```python
cols = ['length','width','height','battery_capacity_mah']
samsung[cols] = samsung[cols].apply(pd.to_numeric, errors='coerce')
```

```python
samsung['volume'] = samsung['length'] * samsung['width'] * samsung['height']
```

```python
samsung['battery_eff']  = (samsung['battery_capacity_mah'] / samsung['volume']) * 100
```

```python
top_balance = samsung.sort_values('battery_eff', ascending=False)[['Name', 'battery_capacity_mah','volume', 'battery_eff']].head(10)
top_balance
```

### Q5. Are flagship devices more battery-efficient per unit size than mid-range ones?


```python
samsung[['ram1_gb','ram2_gb']] = samsung[['ram1_gb','ram2_gb']].apply(
    pd.to_numeric, errors='coerce'
)
```

```python
samsung['segment'] = pd.cut(samsung['ram1_gb'].astype(float), 
                                bins=[0, 4, 8, 100],
                                labels=['Entry', 'Mid','Flagship'])
```

```python
samsung.groupby('segment')['battery_eff'].mean()
```

## âš™ï¸ Optimization & Value Analysis


### Q36. Which devices deliver the best RAM per battery capacity ratio?


```python
samsung.assign(ram_to_battery=samsung['ram1_gb'] / samsung['battery_capacity_mah']).sort_values('ram_to_battery', ascending=False)[['Name','ram1_gb','battery_capacity_mah','ram_to_battery']].head(10)
```

### Q37. Which models maximize performance specs within compact form factors?


```python
samsung[(samsung['ram1_gb'] >= 8) & (samsung['volume'] < samsung['volume'].quantile(0.25))][['Name','ram1_gb','battery_capacity_mah','volume']].head(10)
```

### Q38. Are there devices that appear over-specced for their size?


```python
samsung[(samsung['ram1_gb'] >= samsung['ram1_gb'].quantile(0.9)) & (samsung['volume'] <= samsung['volume'].quantile(0.25))][['Name','ram1_gb','volume']].head(10)
```

### Q39. Which devices offer the best overall spec balance?


```python
samsung['spec_score'] = (samsung['ram1_gb'].rank(pct=True) +
                         samsung['cam1_mp'].rank(pct=True) +
                         samsung['screen_size'].rank(pct=True) +
                         samsung['battery_capacity_mah'].rank(pct=True))
```

```python
samsung.sort_values('spec_score', ascending=False)[['Name','spec_score']].head(10)
```
